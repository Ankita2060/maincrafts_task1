```python
import pandas as pd
```

```python
df=pd.read_csv("student-mat.csv",sep=';')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
duplicates_count = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates_count}")
```

```python
avg_g3 = df['G3'].mean()
print(f"Average Final Grade (G3): {avg_g3:.2f}")
```

```python
above_15 = df[df['G3'] > 15].shape[0]
print(f"Students with G3 > 15: {above_15}")
```

```python
correlation = df['studytime'].corr(df['G3'])
print(f"Correlation between studytime and G3: {correlation:.4f}")
```

```python
gender_performance = df.groupby('sex')['G3'].mean()
print("\nAverage performance by gender:")
print(gender_performance)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.histplot(df['G3'], kde=True)
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade')
plt.ylabel('Frequency')

plt.show()
```

```python
sns.scatterplot(x='studytime', y='G3', data=df, alpha=0.5)
plt.title('Study Time vs Final Grades')
plt.xlabel('Study Time')
plt.ylabel('Final Grade (G3)')

plt.show()
```

```python
gender_avg = df.groupby('sex')['G3'].mean().reset_index().sort_values('G3', ascending=False)
sns.barplot(x='sex', y='G3', data=gender_avg)
plt.title('Average Final Grade (G3) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average grade')
plt.show()
```
